```python
"""
ÁSZF Readability - Regression Approach

Since clustering analysis showed:
1. Clear ordinal structure in embeddings (C1 far from C5, close to C2)
2. No distinct clusters (classes overlap on a continuum)

We'll treat this as REGRESSION instead of classification:
- Predict readability as continuous score (1.0 - 5.0)
- Round to nearest integer for final class
- Use MSE/MAE loss instead of cross-entropy
- Evaluate with both regression metrics AND classification accuracy
"""

from transformers import AutoTokenizer, AutoModel, AutoModelForSequenceClassification
from transformers import get_linear_schedule_with_warmup
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import classification_report, confusion_matrix, mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from torch.optim import AdamW
from pathlib import Path
from tqdm import tqdm
import seaborn as sns
import torch.nn as nn
import pandas as pd
import numpy as np
import warnings
import datetime
import torch

warnings.filterwarnings('ignore')
sns.set_style("whitegrid")

print("=" * 80)
print("ÁSZF READABILITY - REGRESSION APPROACH")
print("=" * 80)

# =============================================================================
# CONFIGURATION
# =============================================================================
model_name = 'SZTAKI-HLT/hubert-base-cc'
BATCH_SIZE = 16
LEARNING_RATE = 1e-5
NUM_EPOCHS = 10
ACCUMULATION_STEPS = 2
MAX_LENGTH = 256
EARLY_STOPPING_PATIENCE = 3

HIDDEN_DROPOUT = 0.2
ATTENTION_DROPOUT = 0.2
FREEZE_LAYERS = 4

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\nDevice: {device}")

torch.manual_seed(42)
np.random.seed(42)


# =============================================================================
# CUSTOM REGRESSION MODEL
# =============================================================================
class BertForReadabilityRegression(nn.Module):
    """
    BERT-based regression model for readability scoring.
    Outputs a single continuous value (1.0 - 5.0).
    """
    def __init__(self, model_name, hidden_dropout=0.2, freeze_layers=4):
        super().__init__()

        # Load pretrained BERT
        self.bert = AutoModel.from_pretrained(
            model_name,
            hidden_dropout_prob=hidden_dropout,
            attention_probs_dropout_prob=hidden_dropout
        )

        # Freeze early layers
        for param in self.bert.embeddings.parameters():
            param.requires_grad = False
        for layer_idx in range(freeze_layers):
            for param in self.bert.encoder.layer[layer_idx].parameters():
                param.requires_grad = False

        hidden_size = self.bert.config.hidden_size

        # Regression head
        self.regressor = nn.Sequential(
            nn.Dropout(hidden_dropout),
            nn.Linear(hidden_size, 256),
            nn.ReLU(),
            nn.Dropout(hidden_dropout),
            nn.Linear(256, 1)
        )

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = outputs.last_hidden_state[:, 0, :]  # [CLS] token
        score = self.regressor(cls_output)
        return score.squeeze(-1)  # Shape: (batch_size,)


# =============================================================================
# DATASET
# =============================================================================
class ASZFRegressionDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        # Keep labels as float (1.0 - 5.0) for regression
        self.labels = [float(l) for l in labels]

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {
            'input_ids': torch.tensor(self.encodings['input_ids'][idx], dtype=torch.long),
            'attention_mask': torch.tensor(self.encodings['attention_mask'][idx], dtype=torch.long),
            'labels': torch.tensor(self.labels[idx], dtype=torch.float)
        }


# =============================================================================
# DATA LOADING
# =============================================================================
print("\n" + "=" * 80)
print("LOADING AND PREPROCESSING DATA")
print("=" * 80)

train_path = Path('/content/sample_data/train.csv')
test_path = Path('/content/sample_data/test.csv')

df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

print(f"\nTraining data: {len(df_train)} samples")
print(f"Test data: {len(df_test)} samples")

# Tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_texts(texts, tokenizer, max_length):
    return tokenizer(
        texts.tolist(),
        max_length=max_length,
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        return_tensors=None
    )

# Tokenize
train_encodings = tokenize_texts(df_train['text'], tokenizer, MAX_LENGTH)
test_encodings = tokenize_texts(df_test['text'], tokenizer, MAX_LENGTH)

# Train/val split
train_indices, val_indices = train_test_split(
    range(len(df_train)),
    test_size=0.2,
    random_state=42,
    stratify=df_train['label_numeric']
)

# Create split encodings
train_enc = {
    'input_ids': [train_encodings['input_ids'][i] for i in train_indices],
    'attention_mask': [train_encodings['attention_mask'][i] for i in train_indices]
}
train_labels = [df_train.iloc[i]['label_numeric'] for i in train_indices]

val_enc = {
    'input_ids': [train_encodings['input_ids'][i] for i in val_indices],
    'attention_mask': [train_encodings['attention_mask'][i] for i in val_indices]
}
val_labels = [df_train.iloc[i]['label_numeric'] for i in val_indices]

test_enc = {
    'input_ids': test_encodings['input_ids'],
    'attention_mask': test_encodings['attention_mask']
}
test_labels = df_test['label_numeric'].tolist()

print(f"\nSplit sizes:")
print(f"  Train: {len(train_labels)}")
print(f"  Val:   {len(val_labels)}")
print(f"  Test:  {len(test_labels)}")

# Datasets
train_dataset = ASZFRegressionDataset(train_enc, train_labels)
val_dataset = ASZFRegressionDataset(val_enc, val_labels)
test_dataset = ASZFRegressionDataset(test_enc, test_labels)

# DataLoaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)


# =============================================================================
# MODEL SETUP
# =============================================================================
print("\n" + "=" * 80)
print("MODEL SETUP")
print("=" * 80)

model = BertForReadabilityRegression(
    model_name=model_name,
    hidden_dropout=HIDDEN_DROPOUT,
    freeze_layers=FREEZE_LAYERS
)
model.to(device)

total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nTotal parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")
print(f"Frozen: {(total_params - trainable_params) / total_params * 100:.1f}%")

# Loss function: MSE for regression
criterion = nn.MSELoss()

# Also track MAE
def compute_mae(preds, targets):
    return np.abs(preds - targets).mean()

# Optimizer
no_decay = ['bias', 'LayerNorm.weight']
optimizer_grouped_parameters = [
    {
        'params': [p for n, p in model.named_parameters()
                   if not any(nd in n for nd in no_decay) and p.requires_grad],
        'weight_decay': 0.01
    },
    {
        'params': [p for n, p in model.named_parameters()
                   if any(nd in n for nd in no_decay) and p.requires_grad],
        'weight_decay': 0.0
    }
]
optimizer = AdamW(optimizer_grouped_parameters, lr=LEARNING_RATE)

total_steps = len(train_loader) * NUM_EPOCHS // ACCUMULATION_STEPS
warmup_steps = int(0.1 * total_steps)
scheduler = get_linear_schedule_with_warmup(optimizer, warmup_steps, total_steps)

print(f"\nLoss: MSE (regression)")
print(f"Optimizer: AdamW (lr={LEARNING_RATE})")


# =============================================================================
# TRAINING FUNCTIONS
# =============================================================================
def train_epoch(model, dataloader, optimizer, scheduler, criterion, device, accum_steps):
    model.train()
    total_loss = 0
    all_preds = []
    all_labels = []
    optimizer.zero_grad()

    progress_bar = tqdm(dataloader, desc='Training')

    for batch_idx, batch in enumerate(progress_bar):
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        scores = model(input_ids, attention_mask)
        loss = criterion(scores, labels) / accum_steps
        loss.backward()

        if (batch_idx + 1) % accum_steps == 0:
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            scheduler.step()
            optimizer.zero_grad()

        total_loss += loss.item() * accum_steps
        all_preds.extend(scores.detach().cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

        progress_bar.set_postfix({'loss': f'{loss.item() * accum_steps:.4f}'})

    # Handle remaining gradients
    if len(dataloader) % accum_steps != 0:
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()

    all_preds = np.array(all_preds)
    all_labels = np.array(all_labels)

    avg_loss = total_loss / len(dataloader)
    mae = compute_mae(all_preds, all_labels)

    # Classification accuracy (round to nearest int)
    pred_classes = np.clip(np.round(all_preds), 1, 5).astype(int)
    accuracy = (pred_classes == all_labels.astype(int)).mean()

    return avg_loss, mae, accuracy


def evaluate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in tqdm(dataloader, desc='Evaluating'):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            scores = model(input_ids, attention_mask)
            loss = criterion(scores, labels)

            total_loss += loss.item()
            all_preds.extend(scores.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    all_preds = np.array(all_preds)
    all_labels = np.array(all_labels)

    avg_loss = total_loss / len(dataloader)
    mae = compute_mae(all_preds, all_labels)

    # Classification metrics
    pred_classes = np.clip(np.round(all_preds), 1, 5).astype(int)
    true_classes = all_labels.astype(int)
    accuracy = (pred_classes == true_classes).mean()

    return avg_loss, mae, accuracy, all_preds, all_labels, pred_classes, true_classes


# =============================================================================
# TRAINING LOOP
# =============================================================================
print("\n" + "=" * 80)
print("TRAINING")
print("=" * 80)

best_val_loss = float('inf')
best_epoch = 0
patience_counter = 0
history = {'train_loss': [], 'train_mae': [], 'train_acc': [],
           'val_loss': [], 'val_mae': [], 'val_acc': []}

model_dir = Path('../models')
model_dir.mkdir(parents=True, exist_ok=True)
best_model_path = model_dir / 'model_regression_best.pt'

for epoch in range(1, NUM_EPOCHS + 1):
    print(f"\n{'=' * 60}")
    print(f"EPOCH {epoch}/{NUM_EPOCHS}")
    print(f"{'=' * 60}")

    train_loss, train_mae, train_acc = train_epoch(
        model, train_loader, optimizer, scheduler, criterion, device, ACCUMULATION_STEPS
    )
    print(f"\nTrain - Loss: {train_loss:.4f}, MAE: {train_mae:.3f}, Acc: {train_acc:.4f}")

    val_loss, val_mae, val_acc, _, _, val_preds, val_true = evaluate(
        model, val_loader, criterion, device
    )
    print(f"Val   - Loss: {val_loss:.4f}, MAE: {val_mae:.3f}, Acc: {val_acc:.4f}")

    # Per-class accuracy
    print("  Per-class recall:", end=" ")
    for c in range(1, 6):
        mask = val_true == c
        if mask.sum() > 0:
            recall = (val_preds[mask] == c).sum() / mask.sum()
            print(f"C{c}:{recall:.2f}", end=" ")
    print()

    history['train_loss'].append(train_loss)
    history['train_mae'].append(train_mae)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_mae'].append(val_mae)
    history['val_acc'].append(val_acc)

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_epoch = epoch
        patience_counter = 0
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'val_loss': val_loss,
            'val_mae': val_mae,
            'val_acc': val_acc
        }, best_model_path)
        print(f"✓ Best model saved!")
    else:
        patience_counter += 1
        print(f"  No improvement. Patience: {patience_counter}/{EARLY_STOPPING_PATIENCE}")
        if patience_counter >= EARLY_STOPPING_PATIENCE:
            print(f"\n⚠ Early stopping at epoch {epoch}")
            break

print(f"\n{'=' * 60}")
print(f"Training complete! Best epoch: {best_epoch}")


# =============================================================================
# FINAL EVALUATION
# =============================================================================
print("\n" + "=" * 80)
print("FINAL EVALUATION")
print("=" * 80)

checkpoint = torch.load(best_model_path, weights_only=False)
model.load_state_dict(checkpoint['model_state_dict'])
print(f"\n✓ Loaded best model from epoch {checkpoint['epoch']}")

# Validation set
val_loss, val_mae, val_acc, val_scores, val_labels, val_pred_classes, val_true_classes = evaluate(
    model, val_loader, criterion, device
)

print(f"\nValidation Results:")
print(f"  MSE Loss: {val_loss:.4f}")
print(f"  MAE: {val_mae:.3f}")
print(f"  Classification Accuracy: {val_acc:.4f}")

print("\nClassification Report (Validation):")
print(classification_report(
    val_true_classes, val_pred_classes,
    target_names=[f'Class {i}' for i in range(1, 6)],
    digits=4
))

# Test set
test_loss, test_mae, test_acc, test_scores, test_labels, test_pred_classes, test_true_classes = evaluate(
    model, test_loader, criterion, device
)

print(f"\nTest Results:")
print(f"  MSE Loss: {test_loss:.4f}")
print(f"  MAE: {test_mae:.3f}")
print(f"  Classification Accuracy: {test_acc:.4f}")

print("\nClassification Report (Test):")
print(classification_report(
    test_true_classes, test_pred_classes,
    target_names=[f'Class {i}' for i in range(1, 6)],
    digits=4
))


# =============================================================================
# VISUALIZATIONS
# =============================================================================
print("\n" + "=" * 80)
print("VISUALIZATIONS")
print("=" * 80)

# 1. Confusion Matrix
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

cm_val = confusion_matrix(val_true_classes, val_pred_classes)
sns.heatmap(cm_val, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=[f'Pred {i}' for i in range(1, 6)],
            yticklabels=[f'True {i}' for i in range(1, 6)])
axes[0].set_title('Validation Set', fontsize=12, fontweight='bold')

cm_test = confusion_matrix(test_true_classes, test_pred_classes)
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Greens', ax=axes[1],
            xticklabels=[f'Pred {i}' for i in range(1, 6)],
            yticklabels=[f'True {i}' for i in range(1, 6)])
axes[1].set_title('Test Set', fontsize=12, fontweight='bold')

plt.suptitle('Regression Model - Confusion Matrices', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(model_dir / 'regression_confusion_matrices.png', dpi=150)
plt.show()

# 2. Predicted vs True Scores
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].scatter(val_labels, val_scores, alpha=0.5, s=20)
axes[0].plot([1, 5], [1, 5], 'r--', label='Perfect prediction')
axes[0].set_xlabel('True Score')
axes[0].set_ylabel('Predicted Score')
axes[0].set_title('Validation: Predicted vs True', fontsize=12, fontweight='bold')
axes[0].legend()
axes[0].set_xlim(0.5, 5.5)
axes[0].set_ylim(0.5, 5.5)

axes[1].scatter(test_labels, test_scores, alpha=0.5, s=20)
axes[1].plot([1, 5], [1, 5], 'r--', label='Perfect prediction')
axes[1].set_xlabel('True Score')
axes[1].set_ylabel('Predicted Score')
axes[1].set_title('Test: Predicted vs True', fontsize=12, fontweight='bold')
axes[1].legend()
axes[1].set_xlim(0.5, 5.5)
axes[1].set_ylim(0.5, 5.5)

plt.suptitle('Regression Model - Score Predictions', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(model_dir / 'regression_scatter.png', dpi=150)
plt.show()

# 3. Score distribution by true class
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

val_df = pd.DataFrame({'true': val_true_classes, 'pred': val_scores})
test_df = pd.DataFrame({'true': test_true_classes, 'pred': test_scores})

val_df.boxplot(column='pred', by='true', ax=axes[0])
axes[0].set_xlabel('True Class')
axes[0].set_ylabel('Predicted Score')
axes[0].set_title('Validation: Predicted Scores by True Class')
axes[0].axhline(y=1.5, color='gray', linestyle=':', alpha=0.5)
axes[0].axhline(y=2.5, color='gray', linestyle=':', alpha=0.5)
axes[0].axhline(y=3.5, color='gray', linestyle=':', alpha=0.5)
axes[0].axhline(y=4.5, color='gray', linestyle=':', alpha=0.5)

test_df.boxplot(column='pred', by='true', ax=axes[1])
axes[1].set_xlabel('True Class')
axes[1].set_ylabel('Predicted Score')
axes[1].set_title('Test: Predicted Scores by True Class')
axes[1].axhline(y=1.5, color='gray', linestyle=':', alpha=0.5)
axes[1].axhline(y=2.5, color='gray', linestyle=':', alpha=0.5)
axes[1].axhline(y=3.5, color='gray', linestyle=':', alpha=0.5)
axes[1].axhline(y=4.5, color='gray', linestyle=':', alpha=0.5)

plt.suptitle('Regression Model - Score Distribution', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(model_dir / 'regression_boxplot.png', dpi=150)
plt.show()

# 4. Training history
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
epochs = range(1, len(history['train_loss']) + 1)

axes[0].plot(epochs, history['train_loss'], 'b-o', label='Train')
axes[0].plot(epochs, history['val_loss'], 'r-o', label='Val')
axes[0].axvline(x=best_epoch, color='g', linestyle='--', alpha=0.7)
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE Loss')
axes[0].set_title('Loss')
axes[0].legend()

axes[1].plot(epochs, history['train_mae'], 'b-o', label='Train')
axes[1].plot(epochs, history['val_mae'], 'r-o', label='Val')
axes[1].axvline(x=best_epoch, color='g', linestyle='--', alpha=0.7)
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('MAE')
axes[1].set_title('Mean Absolute Error')
axes[1].legend()

axes[2].plot(epochs, history['train_acc'], 'b-o', label='Train')
axes[2].plot(epochs, history['val_acc'], 'r-o', label='Val')
axes[2].axvline(x=best_epoch, color='g', linestyle='--', alpha=0.7)
axes[2].set_xlabel('Epoch')
axes[2].set_ylabel('Accuracy')
axes[2].set_title('Classification Accuracy')
axes[2].legend()

plt.suptitle('Training History', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(model_dir / 'regression_history.png', dpi=150)
plt.show()


# =============================================================================
# WITHIN-1 ACCURACY (ORDINAL METRIC)
# =============================================================================
print("\n" + "=" * 80)
print("ORDINAL EVALUATION METRICS")
print("=" * 80)

def within_k_accuracy(true, pred, k=1):
    """Accuracy where prediction within k classes of true is considered correct."""
    return (np.abs(true - pred) <= k).mean()

print("\nValidation Set:")
print(f"  Exact accuracy:     {(val_pred_classes == val_true_classes).mean():.4f}")
print(f"  Within-1 accuracy:  {within_k_accuracy(val_true_classes, val_pred_classes, 1):.4f}")
print(f"  Within-2 accuracy:  {within_k_accuracy(val_true_classes, val_pred_classes, 2):.4f}")

print("\nTest Set:")
print(f"  Exact accuracy:     {(test_pred_classes == test_true_classes).mean():.4f}")
print(f"  Within-1 accuracy:  {within_k_accuracy(test_true_classes, test_pred_classes, 1):.4f}")
print(f"  Within-2 accuracy:  {within_k_accuracy(test_true_classes, test_pred_classes, 2):.4f}")


# =============================================================================
# SUMMARY
# =============================================================================
print("\n" + "=" * 80)
print("SUMMARY")
print("=" * 80)

print(f"""
REGRESSION APPROACH RESULTS:

Model: {model_name} + Regression Head
Loss: MSE (continuous readability score 1.0-5.0)

Validation:
  - MAE: {val_mae:.3f}
  - Exact Accuracy: {val_acc:.4f}
  - Within-1 Accuracy: {within_k_accuracy(val_true_classes, val_pred_classes, 1):.4f}

Test:
  - MAE: {test_mae:.3f}
  - Exact Accuracy: {test_acc:.4f}
  - Within-1 Accuracy: {within_k_accuracy(test_true_classes, test_pred_classes, 1):.4f}

Key insight: MAE shows average prediction error in class units.
Within-1 accuracy shows how often we're at most 1 class off.

Files saved to: {model_dir}
""")

print("SCRIPT COMPLETE!")
```
